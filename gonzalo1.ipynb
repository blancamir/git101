```python
# crea media de samples de una lista (lista = datos, numero= numero samples)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def crear_media_sample(lista,numero=10000):
    """ devuelve la media de un numero de samples
    de una lista; lista=seria la lista de donde queremos
    hacer samples; numero= el numero de samples que queremos hacer (IMPORTAR PANDA)"""
    lista_medias=[]
    for i in range(numero):
        sample=np.random.choice(lista,5)
        lista_medias.append(sample.mean())
    return np.mean(lista_medias)
```

```python
#La función en Python que calcula el número de combinaciones (no importa orden no se repiten) queda así:
from math import factorial

def numero_combinaciones(m, n):
    """Calcula y devuelve el número de combinaciones
       posibles que se pueden hacer con m elementos
       tomando n elementos a la vez.
    """
    return factorial(m) // (factorial(n) * factorial(m - n))
```

```python
import collections 
import random   

count=0
for i in range(10000):   
   """ 
   genero una lista radomn con repeticion , cada numero representa una persona/plato 
   esa lista seria un dia en el trabajo, y cuento las veces que se repiten 4 veces la misma persona/        plato
   """
    platos = random.choices(range(1,6), k=5)
    # con collection counter cuento las veces que se repiten los numeros 
    x=collections.Counter(platos)
    valores = x.values()   
    # si un numero se repite 4 veces lo sumo a count
    if 4 in valores:
        count+=1

porcentaje= count/10000
print (f'el porcentaje que se rompan cuatro platos por la misma persona  es {(count/10000)} y el numero de count {count}')
print (f' la probabilidad que matt rompa 4 platos es {porcentaje * 1/6}')
```
